```r
# 加载库
library(MASS)
library(ISLR)
```

```r
# 简单线性回归

names(Boston)
```

```r
lm.fit <- lm(medv~lstat, data = Boston)
summary(lm.fit)
```

```r
names(lm.fit)
```

```r
coef(lm.fit)
```

```r
# 计算置信区间
confint(lm.fit)
```

```r
predict(lm.fit,data.frame(lstat=(c(5,10,15))), interval = "confidence")
```

```r
predict(lm.fit,data.frame(lstat=(c(5,10,15))), interval = "prediction")
```

```r
plot(Boston$lstat,Boston$medv)
abline(lm.fit)
```

```r
par(mfrow = c(2,2))
plot(lm.fit)
```

```r
# 拟合残差
plot(predict(lm.fit),residuals(lm.fit),main = "Residuals Results")
# 学生化残差
plot(predict(lm.fit),rstudent(lm.fit), main = "Rstudent Results")
```

```r
# 计算杠杆统计量
plot(hatvalues(lm.fit))
# 查向量中最大元素的索引
which.max(hatvalues(lm.fit))
```
