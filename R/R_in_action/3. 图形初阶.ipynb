```r
#data 
dose <- c(20,30,40,45,60)
drugA <- c(16,20,27,40,60)
drugB <- c(15,18,25,31,40)

plot(dose,drugA,type = "b")
```

```r
opar <- par(no.readonly=TRUE)  # 可以生成一个可以修改的当前图形参数列表
par(lty = 2,pch = 17)
plot(dose,drugA,type = "b")
par(opar)
```

```r
plot(dose,drugA,type = "b",lty = 3, lwd = 3,pch = 15,cex=2)
```

```r
par(pin = c(2,3),lwd = 2,cex = 1.5,cex.axis = 0.75,font.axis = 3) 
# pin 以英寸表示图形尺寸 宽和高  lwd 指定线条宽度  cex 相对默认大小缩放的倍数 font 指定绘图使用的字体样式 1 常规 2 粗体 3 斜体
plot(dose,drugA,type = "b",pch = 19,lty = 2,col = "red") # lty 指定线条类型  pch 指定绘制点时使用的符号 
plot(dose,drugB,type = "b",pch = 23,lty = 6,col = "blue",bg = "green")
```

```r
plot(dose,drugA,type = "b",col = "red",lty = 2,pch = 2,lwd = 2,
    main = "Clinical Trials for Drug A",
    sub = "This is hypothetical data",
    xlab = "Dosage", ylab = "Drug Respose",
    xlim = c(0,60), ylim = c(0,70))
```

```r
x <- c(1:10)
y <- x
z <- 10/x
opar <- par(no.readonly=TRUE)
par(mar = c(5,4,4,8)+0.1) # mar 以数值向量表示边界大小  单位为英分 顺序为“下、左、上、右” 
plot(x,y,type = "b",pch = 21, col = "red", yaxt = "n",lty = 3, ann = FALSE)
# yaxt="n" 禁用Y轴 会保留下框线，只是去除了刻度 ann=FALSE 移除默认的标题和标签
lines(x,z,type = "b",pch = 22, col = "blue", lty = 2) # lines 为一幅现有图形添加新的图形元素
axis(side = 2,at = x, labels = x, col.axis = "red", las = 2) 
# side 表示在图形的哪边绘制坐标轴 1234 下左上右  at 表示需要绘制刻度线的位置  lables 表示刻度线旁边的文字标签 
# las 标签是否平行于（=0）或垂直（=2）于坐标轴
axis(4,at = z, labels = round(z,digits = 2),col.axis = "blue", las = 2, cex.axis = 0.7, tck = -0.01) 
# tck 刻度线的长度
mtext("y=1/x", side = 4, line = 3, cex.lab = 1, las = 2, col = "blue")
# mtext 用于在图形的边界添加文本  text() 用于在图像内部添加文本 
title("An Example of Creative Axes", xlab = "X values", ylab = "Y=X")
par(opar)
```

```r
opar <- par(no.readonly = TRUE)
par(lwd = 2, cex = 1.5, font.lab = 2)
plot(dose, drugA, type = "b", pch = 15, lty = 1, col = "red", ylim = c(0,60), 
    main = "Drug A vs Drug B", xlab = "Drug Dosage", ylab = "Drug Response")
lines(dose, drugB, type = "b", pch = 17, lty = 2, col = "blue")
abline(h = c(30), lwd = 1.5, lty = 2, col = "gray")
# abline 用来为图形添加参考线 h = yvalues, v = xvalues
library(Hmisc)
minor.tick(nx = 3, ny = 3, tick.ratio = 0.5)
# minor.tick() 创建次要刻度线  需要Hmisc包 nx ny 分别指定了x轴和y轴每两条主刻度线之间通过次要刻度线划分得到的区间个数
# tick.ratio 表示次要刻度线相对于主刻度线的大小比例
legend("topleft", inset = 0.05, title = "Drug Type", c("A","B"), 
       lty = c(1,2), pch = c(15,17), col = c("red","blue"))
par(opar)
```
