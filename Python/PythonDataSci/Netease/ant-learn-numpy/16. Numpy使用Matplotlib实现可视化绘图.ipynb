## Numpy使用Matplotlib实现可视化绘图

可以直接将Numpy的数组传给Matplotlib实现可视化绘图：  
1. 曲线图
2. 饼图
3. 柱状图
4. 直方图

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1. 绘制正弦曲线

```python
x = np.linspace(-10, 10, 100)
x.shape
```

```python
y = np.sin(x)
y.shape
```

```python
plt.plot(x, y)
plt.show()
```

### 2. 绘制饼图

```python
# 数字array
arr = np.array([15, 30, 45, 10])
```

```python
# 名字
labels = np.array(['Frogs', 'Hogs', 'Dogs', 'Logs'])
```

```python
plt.pie(arr, labels=labels, autopct='%1.1f%%')
plt.show()
```

### 3. 柱状图

```python
# 数字array
arr = np.array([15, 30, 45, 10])
```

```python
# 数字名字
labels = np.array(['Frogs', 'Hogs', 'Dogs', 'Logs'])
```

```python
plt.bar(labels, arr)
plt.show()
```

### 4. 直方图

```python
# 采样得到正态分布的随机数10000个
mu, sigma = 2, 1
arr = np.random.normal(mu, sigma, 10000)
```

```python
arr
```

```python
plt.hist(arr, bins=100, density=1)
plt.show()
```
