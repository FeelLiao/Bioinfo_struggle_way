## 1.2 matplotlib 的基础绘图框架

```python
# 最基本的绘图程序框架
from matplotlib import pyplot as plt
plt.plot([1,2,2,1])
plt.show()
```

### 1.2.1 创建Figure对象
    Figure对象是后续绘图的总容器

    matplotlib.pyplot.figure(
        num : 图形序号（ID）不指定时自动递增，如果图形已经存在，则激活相应图形
        figsize
        dpi
        facecolor : 图形背景色
        edgecolor : 图形边框色
        frameon=True : 是否绘制图形外框架
        FigureClass : 使用自定义的matplotlib.figure.Figure类
        clear=False : 图形已存在时是否清楚原图像
        )

### 1.2.2 调用plot() 在Figure对象中绘图
    plot()是在Axes（子图）对象上绘图 一个Figure对象可以有多个Axes对象
    
    matplotlib.pyplot.plot(
        x,y 
        data : 变量列所对应的数据框名称
        fmt: 对图像进行格式设置
        )

```python
import numpy as np 
t=np.linspace(-3,3,100)
x=16*(np.sin(t))**3
y=13*np.cos(t)-5*np.cos(2*t)-2*np.cos(3*t)-np.cos(4*t)
a=plt.figure(figsize=(5,5))
plt.plot(x,y)
```

### 1.2.4 设置当前Axes对象的属性
    xlim/ylim : 设置x,y轴的显示刻度范围
    xlabel/ylabel : 设置x,y轴的标题文字
    clabel : 为轮廓线设置标签文字
    title : 设置子图的标题
    legend : 显示图例
    annotate : 绘制图形标注
    axhspan : 绘制垂直或水平色块
    fill : 填充区域
