```python
# 导入ccss数据
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
ccss_path="./PythonData/CCSS_sample.xlsx"
ccss_data=pd.read_excel(ccss_path,sheet_name=0)
```

# 3.2 饼图

```python
ccss_data["s5"].value_counts().plot.pie()
```

```python
plt.pie(ccss_data.s5.value_counts(),labels=ccss_data.s5.value_counts().index)
```

```python
plt.pie(x=ccss_data.s5.value_counts(),labels=ccss_data.s5.value_counts().index,
       explode=[0,0,0.5,0,0,0,0,0,0,0,0])

## shadow=True , startangle=90, radius=2, counterclock=False, rotatelabels=True
```

```python
# 半圆图 
# 用plt.pie()默认数据达到100% 就会绘制饼图，所以只达到50%即可绘制半圆图
plt.pie(x=ccss_data.s5.value_counts(normalize=True)/2,labels=ccss_data.s5.value_counts().index)
# normalize=True   数据标准化  计算数据百分比
```

```python
# 圆环图 
# 在同一个Axes上绘制两个半径不一样的饼图 其中一个为白底
plt.figure(figsize=(6,6))
plt.pie(x=ccss_data.s5.value_counts(),labels=ccss_data.s5.value_counts().index)
plt.pie([1],colors=['#ffffff'],radius=0.7)
# pie 默认的radius=1
```

# 3.3 条带图

    seaborn.stripplot(
     x,y,hue : 分类，连续变量，色彩分组变量名
     data ：数据框名称
     order : hue变量各类别取值的绘图顺序
     orient : 'v' 'h' 条带绘制方向
     jitter=True 是否对散点位置进行随机抖动以避免重叠
     color
     palette
     size
     linewidth
     ax
     )

```python
# 随机抖动关闭 样本较多时散点会部分重叠，使图形失去使用价值
sns.stripplot(y=ccss_data.s3[:200],jitter=False)
```

```python
# 随机抖动开启
sns.stripplot(y=ccss_data.s3[:200],jitter=1)
# jitter 默认为开启
```

```python
sns.swarmplot(y=ccss_data.s3[:200])
```

# 3.4直方图、KDE图与地毯图

```python
ccss_data.s3.plot.hist()
```

```python
plt.hist(ccss_data.s3)
```

    用seaborn实现
    seaborn.distplot(
        a: 绘图用的数据列，有变量名称会被用作轴标签
        bins=None 组段数
        
        hist=True
        kde=True gaussian 核密度估计曲线
        rug =False rugplot
        
        vertical=False
        norm_hist=False 纵轴显示概率密度而不是原始频数
        
        ）

```python
# 只绘制rugplot
sns.distplot(ccss_data.s3,hist=False,kde=False,rug=True)
```

```python
# 只绘制直方图
sns.distplot(ccss_data.s3,kde=False)
```

```python
# 默认图形输出
sns.distplot(ccss_data.s3)
```

```python
# 和正态分布曲线比较
from scipy.stats import norm
sns.distplot(ccss_data.s3,fit=norm)
```

```python
sns.distplot(ccss_data.s3,vertical=True)
```

```python
# 自定义图像细节
sns.distplot(ccss_data.s3,rug=True,
             rug_kws={'color':"g"},
             kde_kws={"color":"k","lw":3,"label":"KDE"},
             hist_kws={"lw":3,"color":"y","label":"Hist"})
```

```python
# kdeplot
# 使用了Axes叠加绘图的功能
sns.kdeplot(ccss_data.s3,bw=0.05,label="bw:0.05")
sns.kdeplot(ccss_data.s3,bw=0.1,label="bw:0.1")
sns.kdeplot(ccss_data.s3,bw=0.2,label="bw:0.2")
```

# 3.5箱图

```python
ccss_data.s3.plot.box()
```

```python
plt.boxplot(ccss_data.s3)
```

    用seaborn 实现
        seaborn.boxplot(
        x,y,hue : 分类，连续变量，色彩分组变量名
        data ：数据框名称
        order : hue变量各类别取值的绘图顺序
        orient : 'v' 'h' 条带绘制方向
        saturation=0.75 : float 直条颜色饱和度
        palette
        width=0.8 箱体宽度所占比例
        fliersize=5 离群值散点大小
        linewidth=None 框线宽度
        whis=1.5 离群值确定标准，距离IQR上下界的倍数
        ）

```python
sns.boxplot(y=ccss_data.s3)
```

```python
sns.boxplot(y=ccss_data.index1,saturation=0.3,width=0.5,linewidth=3,fliersize=10)
```

### 增强箱图
    seaborn.boxenplot(
        x,y,hue
        
        k_depth='proportion'
        
            "proportion" | "tukey" | "trustworthy"
        scale="exponential" 箱体宽度和样本量的对应关系
            "linear"  等比例减少
            "exponential" 对应箱体外侧未覆盖的样本比例
            "area" 对应箱体本身
        outlier_prop=0.007 离群值在样本中的比例

```python
sns.boxenplot(x=ccss_data.s2,y=ccss_data.index1)
```

```python
sns.boxenplot(x=ccss_data.time,y=ccss_data.index1)
```

```python
sns.boxenplot(x=ccss_data.time,y=ccss_data.index1,outlier_prop=0.1)
```

### 提琴图

    seaborn.violinplot(
        x,y,hue,data,order
        bw="scott" : KDE图形的具体计算方法或核宽度
            "scott" | "silverman"
        cut=2 绘图时是否超越两端的极值 0 严格限定
        scale="area" 各KDE图形设置为相同面积/按照频数设置宽度/相同宽度
            "area" | "count" | "width" 
        gridsize=100 用于形成KDE图形的散点总数
        width=0.8 图形占的宽度比例
        inner="box" 提琴内部呈现的图形种类 箱图/分位数/原始数据
            "box" | "quartile" | "point" | "stick" | None
        saturation=0.75 图形色彩饱和度
        ）

```python
sns.violinplot(y=ccss_data.s3)
```

```python
sns.violinplot(y=ccss_data.s3,inner="quartile")
```

```python
sns.violinplot(y=ccss_data.s3,inner=None)
sns.swarmplot(y=ccss_data.s3,size=2,color="k")
```
