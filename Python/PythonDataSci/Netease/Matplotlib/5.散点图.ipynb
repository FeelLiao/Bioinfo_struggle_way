# 5.1 普通散点图

```python
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
ccss_path="./PythonData/CCSS_sample.xlsx"
ccss_data=pd.read_excel(ccss_path,sheet_name=0)
```

```python
# 用matplotlib 实现
ccss_data.plot.scatter("s3","index1")
```

```python
plt.scatter(x=ccss_data.s3,y=ccss_data.index1,s=ccss_data.index1,c=ccss_data.index1)
```

```python
# seaborn 实现 
# seaborn.scatterplot()  和lineplot非常类似
sns.scatterplot(x=ccss_data.s3,y=ccss_data.index1,hue=ccss_data.s2)
```

# 5.2 变量间的回归趋势考察

```python
# seaborn.lmplot() 考察各种常见的线性，曲线回归趋势，可按照分类变量分组/分行列面板
# 并且同时计算可行区间
```

```python
sns.lmplot(x="s3",y="index1",data=ccss_data)
```

```python
# 对年龄进行分段散点图呈现，分段后呈现因变量CI而不是原始数值
sns.lmplot(x="s3",y="index1",data=ccss_data,x_bins=15)
```

```python
# 不分段呈现因变量CI 而不是原始数值
sns.lmplot(x="s3",y="index1",data=ccss_data,x_estimator=np.mean)
```

```python
sns.lmplot(x="s3",y="index1",data=ccss_data,x_estimator=np.median,x_bins=15)
```
