# 7.2 分区绘制多个子图

## 7.2.1 subplot命令

```python
# matplotlib.pyplot.subplot()  nrows/ncols 图组的行/列数  index 指定所创建的Axes对象所在区域
# subplot(323)  和 subplot(3,2,3) 含义相同
plt.subplot(2,2,1) # 第一行的左图
plt.subplot(2,2,2) # 第一行的右图
plt.subplot(2,1,2) # 两行一列  上面两个图，下面一个图，第二整行
plt.plot([1,2,2,1]) # 在最新创建的子图上绘图
plt.show()
```

## 7.2.2 切换使用的子图对象

```python
# matplotlib.pyplot.sca(ax) 子图对象所在的图表对象也会被自动设定为当前图表对象
plt.subplot(221)
ax1=plt.subplot(222)
plt.subplot(2,1,2) # 两行一列  上面两个图，下面一个图，第二整行
plt.sca(ax1)
plt.plot([1,2,2,1]) # 在最新创建的子图上绘图
plt.show()
```

```python
# 多figure对象，多子图对象的联合应用
plt.figure(1)
plt.figure(2)
ax1=plt.subplot(211) # 在图表2中创建子图1
ax2=plt.subplot(212) # 在图表2中创建子图2
x=np.linspace(0,3,100)
for i in range(5):
    plt.figure(1)
    plt.plot(x,np.exp(i*x/3))
    plt.sca(ax1)
    plt.plot(x,np.sin(i*x))
    plt.sca(ax2)
    plt.plot(x,np.cos(i*x))
plt.show()
```

```python
# 使用数组方式指定子图
fig,axes=plt.subplots(1,2,sharey=True)
sns.boxplot(y=ccss_data.index1a,ax=axes[0])
sns.boxplot(y=ccss_data.index1b,ax=axes[1])
```

# 7.3 不等距复杂网格的设定

## 7.3.1 subplot2grid方法

    matplotlib.pyplot.subplot2grid(shape,loc,rowspan,colspan)
        shape 设定整体网格行列数
        loc 指定网格起点  rowspan/colspan 指定网格跨度

```python
plt.figure()
ax1=plt.subplot2grid(shape=(3,3),loc=(0,0),colspan=3)
ax2=plt.subplot2grid(shape=(3,3),loc=(1,0),colspan=2)
ax5=plt.subplot2grid((3,3),(2,1))
```

## 7.3.2 GridSpec方法

    matplotlib.gridspec.GridSpec(
        nrows/ncols
        figure
        legt/bottom/right/top
        wspace/hspace
        width_ratios/height_ratios
        )

```python
import matplotlib.gridspec as gridspec
f=plt.figure()
gs=gridspec.GridSpec(nrows=2,ncols=2,width_ratios=[1,2],height_ratios=[4,1])
# 宽度比 1：2 高度比 4:1
ax1=plt.subplot(gs[0])
ax2=plt.subplot(gs[1])
ax3=plt.subplot(gs[2])
ax4=plt.subplot(gs[3])
plt.show()
```

```python
# 自定义函数，隐藏数轴刻度
def make_ticklabels_invisible(fig):
    for i, ax in enumerate(fig.axes):
        ax.text(0.5, 0.5, "ax%d" % (i+1), va="center", ha="center")
        ax.tick_params(labelbottom=False, labelleft=False)
fig=plt.figure()
gs=gridspec.GridSpec(3,3)
ax1=plt.subplot(gs[0,:])
ax2=plt.subplot(gs[1,:-1])
ax3=plt.subplot(gs[1:,-1])
ax4=plt.subplot(gs[-1,0])
ax5=plt.subplot(gs[-1,-2])
fig.suptitle("$GridSpec$")
make_ticklabels_invisible(fig)
plt.show()
```
