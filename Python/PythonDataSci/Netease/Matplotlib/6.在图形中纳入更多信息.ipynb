# 6.1 设置图例

    matplotlib.pyplot.legend()
        handles: 需要设置图例标签的图形元素列表
        labels： 相应图像元素的图例标签文字列表
        loc 图例的位置

```python
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
ccss_path="./PythonData/CCSS_sample.xlsx"
ccss_data=pd.read_excel(ccss_path,sheet_name=0)
```

```python
# 面板图形中图例的显示
sns.lmplot(x="s3",y="index1",hue="s2",col="s0",data=ccss_data)
```

```python
sns.lmplot(x="s3",y="index1",hue="s2",col="s0",data=ccss_data)
```

# 6.2 混合图形

```python
# matplotlib 中，实现混合图形的原理是多个图形元素直接叠加
ax1=sns.kdeplot(ccss_data.s3)
```

```python
ax1=sns.kdeplot(ccss_data.s3)
sns.distplot(ccss_data.s3,kde=False,norm_hist=True,ax=ax1)
```

```python
ax1=sns.kdeplot(ccss_data.s3)
ax1.hist(ccss_data.s3,density=True)
```

# 6.3双轴图

```python
# matplotlib 实现
plt.gca().twinx()
```

```python
# 设置第一y轴的图像
ccss_data.index1.plot.kde(label="index1")
plt.gca().set_ylabel("$index1$")
# 设置第二y轴的图像
ax2=plt.gca().twinx()
sns.kdeplot(ccss_data.s3,legend=False,ax=ax2)
ax2.set_ylabel("s3")
```

# 6.4 使用行/列面板

```python
# 行/列面板的本质就是指定具体的分组变量用于行列拆分
# seaborn.catplot() 通用的行列面板操作
sns.catplot(x="time",y="index1",data=ccss_data,col="s0",kind="box")
```

```python
sns.catplot(x="time",y="index1",data=ccss_data,col="s0",kind="boxen")
```

```python
sns.catplot(x="time",y="index1",data=ccss_data,col="s0",kind="violin")
```

```python
sns.catplot(x="time",y="index1",data=ccss_data,col="s0",hue="s2",kind="bar")
```

```python
sns.catplot(x="time",y="index1",data=ccss_data,col="s0",hue="s2",kind="bar",col_wrap=2,legend_out=False)
```
